```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('./house_data_set_cleaned_2.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=["Unnamed: 0"], axis=1)
```

```python
df_temp = df[df['Price'] > 1.0e7]
df_temp = df_temp[df_temp['Price'] < 4.0e8]
sns.distplot(np.log(df_temp['Price']), fit=norm)
fig = plt.figure()
res = stats.probplot(np.log(df_temp['Price']), plot=plt)
```

```python
df_temp = df[df['Price'] > 1.0e7]
df_temp = df_temp[df_temp['Price'] < 4.0e8]
sns.distplot(np.log(df_temp['House Size']), fit=norm)
fig = plt.figure()
res = stats.probplot(np.log(df_temp['House Size']), plot=plt)
```

```python
sns.distplot(np.log(df[df['Price'] > 1.0e8]['Land Size']), fit=norm)
fig = plt.figure()
res = stats.probplot(np.log(df[df['Price'] > 1.0e8]['Land Size']), plot=plt)
```

```python
sns.distplot(np.log(df[df['Price'] > 1.0e8]['Beds']), fit=norm)
fig = plt.figure()
res = stats.probplot(np.log(df[df['Price'] > 1.0e8]['Beds']), plot=plt)
```
